```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import networkx as nx
import matplotlib.pyplot as plt

# Download the stopwords and punkt resources from NLTK
nltk.download('stopwords')
nltk.download('punkt')

# Read in the text and tokenize it
with open('dalloway.txt', 'r', encoding='utf-8') as file:
    text = file.read()
tokens = word_tokenize(text)

# Remove stopwords and punctuation
stop_words = set(stopwords.words('english'))
punctuation = [',', '.', ';', ':', '!', '?', '-', '(', ')', '[', ']']
words = [word.lower() for word in tokens if word.lower() not in stop_words and word not in punctuation]

# Find all the character names in the text
characters = []
for i in range(len(words) - 1):
    if words[i].istitle() and words[i + 1].istitle():
        characters.append((words[i], words[i + 1]))

# Create the character network
G = nx.Graph()
G.add_edges_from(characters)

# Draw the character network
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500)
nx.draw_networkx_edges(G, pos, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_family='Verdana', font_weight='bold')
plt.axis('off')
plt.show()
```
